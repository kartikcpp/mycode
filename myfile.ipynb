```python
from sklearn.preprocessing import add_dummy_feature
X_b=add_dummy_feature(x)
X_b
theta_best=np.linalg.inv(X_b.T@X_b)@X_b.T @y
theta_best
```

```python
X_new=[[3],[2]]
xnb=add_dummy_feature(X_new)
yy=xnb@theta_best
```

```python
plt.plot(X_new,yy,"b-",label="pred")
plt.plot(x,y,"r.")
plt.show()
```
